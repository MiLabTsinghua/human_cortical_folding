```r
library(dplyr)
library(ComplexUpset)
library(ggplot2)
```

```r
setwd('~/workspace/Mida_collab/')
```

```r
res_mat_bi <- read.csv('CellOracle/output/GSEA_mat_bi_exp0.04pct0.05size20.csv',header = TRUE,row.names = 1)
colnames(res_mat_bi) <- c('4','5','3','2')
```

```r
res_mat_bi  %>% head()
```

```r
# 'Outside of known sets',
g <- upset(res_mat_bi, intersect = c("4", "5", "3","2"),
           intersections = list(c('2','5'),c('3','2'),c('4','5'),'2','3','5','4'), 
           theme=theme_grey(),sort_sets=FALSE,sort_intersections=FALSE)
```

```r
g
```

```r
set <- g[[4]]+coord_flip()
set
```

```r
intersection_bar <- g[[2]]+coord_flip()+scale_y_reverse()
intersection_bar
```

```r
set_bars <- g[[3]]+ coord_cartesian()
set_bars+ scale_y_continuous()
```

```r
ggsave(set_bars+ scale_y_continuous()+scale_x_discrete(labels = c('4','5','3','2')), file='CellOracle/output/GSEA_set_barplots_final.pdf' )
```

```r
ggsave(intersection_bar,file='CellOracle/output/GSEA_intersection_barplot_final.pdf')
```

```r
ggsave(set,file = 'CellOracle/output/GSEA_setplot_final.pdf')
```
